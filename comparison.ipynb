```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

old_data = pd.read_csv('sample_submission.csv', index_col="Id")
new_data = pd.read_csv('submission.csv', index_col="Id")
```

```python
sns.distplot(a=old_data['SalePrice'], label="Old_Sale", kde=False)
sns.distplot(a=new_data['SalePrice'], label="New_Sale", kde=False)
plt.legend()
plt.show();
```
